```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# -------------------------------
# 1. Normal Distribution
# -------------------------------

# Generate data from normal distribution
mean = 0
std_dev = 1
data = np.random.normal(mean, std_dev, 1000)

# Plot histogram
plt.hist(data, bins=30, density=True, alpha=0.6, color='skyblue')

# Plot PDF using SciPy
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, mean, std_dev)
plt.plot(x, p, 'k', linewidth=2)

plt.title("Normal Distribution (Mean=0, Std Dev=1)")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.show()

# -------------------------------
# 2. Descriptive Statistics
# -------------------------------

print("Descriptive Statistics:")
print("Mean:", np.mean(data))
print("Median:", np.median(data))
print("Variance:", np.var(data))
print("Standard Deviation:", np.std(data))

# -------------------------------
# 3. Coin Toss Simulation
# -------------------------------

# Simulate 1000 coin tosses
tosses = np.random.choice(['H', 'T'], size=1000)

# Count occurrences
heads = np.sum(tosses == 'H')
tails = np.sum(tosses == 'T')

print("\nCoin Toss Simulation (1000 tosses):")
print("Heads:", heads)
print("Tails:", tails)
print("Probability of Heads:", heads/1000)
print("Probability of Tails:", tails/1000)

# Bar graph for coin toss result
plt.bar(['Heads', 'Tails'], [heads, tails], color=['blue', 'orange'])
plt.title("Coin Toss Simulation Results")
plt.ylabel("Count")
plt.show()
```
