```python
from collections import deque

graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A'],
    'D': ['B'],
    'E': ['B']
}

# Breadth First Search
def bfs(graph, start):
    visited = set()
    queue = deque([start])
    visited.add(start)

    print("BFS Traversal:")
    while queue:
        node = queue.popleft()
        print(node, end=" ")
        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)
    print()

# Depth First Search
def dfs(graph, start, visited=None):
    if visited is None:
        visited = set()
        print("DFS Traversal:")

    visited.add(start)
    print(start, end=" ")

    for neighbor in graph[start]:
        if neighbor not in visited:
            dfs(graph, neighbor, visited)

# Run
bfs(graph, 'A')
dfs(graph, 'A')
```

```python
import heapq

graph = {
    'A': [('B', 1), ('C', 3)],
    'B': [('D', 3), ('E', 1)],
    'C': [('F', 5)],
    'D': [('G', 1)],
    'E': [('G', 5)],
    'F': [('G', 2)],
    'G': []
}

heuristic = {
    'A': 7,
    'B': 6,
    'C': 4,
    'D': 1,
    'E': 3,
    'F': 2,
    'G': 0
}

def a_star(start, goal):
    pq = []
    heapq.heappush(pq, (heuristic[start], 0, start, [start]))

    while pq:
        f, g, node, path = heapq.heappop(pq)

        if node == goal:
            return path, g

        for neighbor, cost in graph[node]:
            new_g = g + cost
            new_f = new_g + heuristic[neighbor]
            heapq.heappush(pq, (new_f, new_g, neighbor, path + [neighbor]))

# Run
path, cost = a_star('A', 'G')
print("A* Path:", path)
print("Total Cost:", cost)
```

```python
def evaluate(board):
    wins = [(0,1,2),(3,4,5),(6,7,8),
            (0,3,6),(1,4,7),(2,5,8),
            (0,4,8),(2,4,6)]

    for i,j,k in wins:
        if board[i] == board[j] == board[k]:
            if board[i] == 'X':
                return 1
            elif board[i] == 'O':
                return -1
    return 0

def minimax(board, depth, is_max):
    score = evaluate(board)
    if score != 0 or ' ' not in board:
        return score

    if is_max:
        best = -1000
        for i in range(9):
            if board[i] == ' ':
                board[i] = 'X'
                best = max(best, minimax(board, depth+1, False))
                board[i] = ' '
        return best
    else:
        best = 1000
        for i in range(9):
            if board[i] == ' ':
                board[i] = 'O'
                best = min(best, minimax(board, depth+1, True))
                board[i] = ' '
        return best

# Example Board
board = ['X', 'O', 'X',
         'O', 'X', ' ',
         ' ', ' ', 'O']

print("Minimax Value:", minimax(board, 0, True))
```
