```python
# Experiment 05
# Data Manipulation using Iris Dataset (No CSV Required)

# Step 1: Import Libraries
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris

# Step 2: Load Iris Dataset from sklearn
iris = load_iris()

# Convert to DataFrame
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Add Species column
df["Species"] = iris.target

# Convert numeric target to actual species names
df["Species"] = df["Species"].map({
    0: "Setosa",
    1: "Versicolor",
    2: "Virginica"
})

# Step 3: Display First 5 Records
print("First 5 Records:")
print(df.head())

# --------------------------------------------------
# Filtering Example
# --------------------------------------------------

# Filter only Setosa flowers
setosa = df[df["Species"] == "Setosa"]
print("\nFiltered Data (Setosa):")
print(setosa.head())

# --------------------------------------------------
# Aggregate Functions on Petal Length
# --------------------------------------------------

print("\nMinimum Petal Length:", df["petal length (cm)"].min())
print("Maximum Petal Length:", df["petal length (cm)"].max())
print("Average Petal Length:", df["petal length (cm)"].mean())

# --------------------------------------------------
# Grouping Feature-wise
# --------------------------------------------------

group_mean = df.groupby("Species").mean()
print("\nMean Values Grouped by Species:")
print(group_mean)

group_minmax = df.groupby("Species")["petal length (cm)"].agg(["min", "max"])
print("\nPetal Length Min & Max by Species:")
print(group_minmax)

# --------------------------------------------------
# NumPy Operations
# --------------------------------------------------

petal_array = np.array(df["petal length (cm)"])

print("\nNumPy Calculations:")
print("Mean:", np.mean(petal_array))
print("Max:", np.max(petal_array))
print("Min:", np.min(petal_array))
```
