```python
num = int(input("Enter number: "))
fact = 1

for i in range(1, num+1):
    fact = fact * i

print("Factorial =", fact)
```

```python
num = int(input("Enter number: "))

if num % 2 == 0:
    print("Even Number")
else:
    print("Odd Number")
```

```python
num = int(input("Enter number: "))
temp = num
rev = 0

while num > 0:
    digit = num % 10
    rev = rev*10 + digit
    num = num // 10

if temp == rev:
    print("Palindrome")
else:
    print("Not Palindrome")
545
```
