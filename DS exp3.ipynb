```python
import numpy as np

# Create a 3x3 matrix
A = np.array([[1, 2, 3],
              [4, 5, 6],
              [7, 8, 10]])

# Compute determinant
det_A = np.linalg.det(A)

print("Matrix A:\n", A)
print("Determinant of A:", det_A)
```

```python
# Compute inverse (only if determinant is not zero)
if det_A != 0:
    inv_A = np.linalg.inv(A)
    print("Inverse of A:\n", inv_A)
else:
    print("Matrix is singular and cannot be inverted.")
```

```python
# Coefficient matrix
C = np.array([[2, 3],
              [4, 5]])

# Constant matrix
D = np.array([10, 20])

# Solve the system
solution = np.linalg.solve(C, D)

print("Solution (x, y):", solution)
```
